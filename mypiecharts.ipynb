```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set_theme(rc = {"figure.figsize": (10, 4)})
```

```python
df = sns.load_dataset("titanic")
```

```python
df.select_dtypes(exclude = 'number').head(3)
```

Pie charts ou camembert


```python
df[['sex','class']].value_counts().plot(kind = 'pie',
                            autopct = '%1.2f%%',
                            title = 'pie chart');  
plt.title('Titre de mon piechart')
```

Generalisation:

- Le code suivant vous permet de creer autant de pie charts que vous souhaitez mais attention, n'utilisez que pour les variables NOMINALES(ayant un nombre reduit de modalites, au plus 6, 7 modalités)

```python
df.select_dtypes(exclude = 'number').head()
```

```python
def mypiechart(df = df, nrows = 3, ncols = 3):
    fig, ax = plt.subplots(nrows, ncols, figsize = (10, 10))
    for i, colname in enumerate(df.select_dtypes(exclude = 'number').columns):
        val = df[colname].value_counts()
        ax.flat[i].pie(val.values, labels = val.index, autopct = '%1.1f%%')
        ax.flat[i].set_title(colname)
        plt.tight_layout()
    plt.show()
```

```python
mypiechart()
```
